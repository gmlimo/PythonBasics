# Matplotlib Single Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 4, 5, 2, 3, 6]
plt.plot(x)
plt.show()
```

```python
import numpy as np

y = np.arange(10)
plt.plot(y)
plt.show()
```

```python
plt.plot(y, [z**2 for z in y]) #we have 0-9 in x axis and 0, 1, 4, 9, 16... (x axis elevated by 2) for y axis
plt.show()
```

```python
plt.plot(y, y**2) #same as before
plt.show()
```
