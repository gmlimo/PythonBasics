# Color Style Makers for Matplotlib

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(5)
y = x

plt.plot(x, y + 1, 'g')
plt.plot(x, y + 0.5, 'y')
plt.plot(x, y, 'r')
plt.plot(x, y - 0.2, 'c')
plt.plot(x, y - 0.4, 'k')
plt.plot(x, y - 0.6, 'm')
plt.plot(x, y - 0.8, 'w')
plt.plot(x, y - 1, 'b')
plt.show()
```

```python
plt.plot(x, y + 1, 'g', label = 'y + 1')
plt.plot(x, y + 0.5, 'y', label = 'y + 0.5')
plt.plot(x, y, 'r', label = 'y')
plt.plot(x, y - 0.2, 'c', label = 'y - 0.2')
plt.plot(x, y - 0.4, 'k', label = 'y - 0.4')
plt.plot(x, y - 0.6, 'm', label = 'y - 0.6')
plt.plot(x, y - 0.8, 'w', label = 'y - 0.8')
plt.plot(x, y - 1, 'b', label = 'y - 1')
plt.legend()
plt.show()
```

```python
plt.plot(x, y + 1, 'g', x, y + 0.5, 'y', x, y, 'r', x, y - 0.2, 'c',
        x, y - 0.4, 'k', x, y - 0.6, 'm', x, y -0.8, 'w', x, y - 1, 'b') #Alternate way to do the same
plt.show()
```

# Line Styles

```python
plt.plot(x, y + 1, '-', x, y + 0.5, '--', x, y, '-.', x, y - 0.2, ':')
plt.show()
```

```python
plt.plot(x, y + 1, '.') #Markers
plt.plot(x, y + 0.5, ',')
plt.plot(x, y, 'o')
plt.plot(x, y - 0.2, '<')
plt.plot(x, y - 0.4, '>')
plt.plot(x, y - 0.6, 'v')
plt.plot(x, y - 0.8, '^')
plt.plot(x, y - 1, '1')
plt.plot(x, y - 2, '2')
plt.show()
```

```python
plt.plot(x, y, 'mo--') #Color, markers and type of line specified
plt.plot(x, y + 1, 'g*-.')
plt.show()
```

```python
plt.plot(x, y, color = 'g', linestyle = '--', linewidth = 1.5, marker = '^', markerfacecolor = 'b',
        markeredgecolor = 'k', markeredgewidth = 1.5, markersize = 5)
plt.grid(True)
plt.show()
```

```python
x = y = np.arange(10) #To have letters in x axis and numbers on y axis
plt.plot(x, y, 'o--')
plt.xticks(range(len(x)), ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
plt.yticks(range(1, 10, 1))
plt.show()
```
