# Grid, Axes and Labels for Matplotlib

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(3) #Four different plots in one
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
print(plt.axis())
plt.show()
```

```python
x = np.arange(3) 
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.axis([0, 2, 0, 8]) #limit the axis range (0-2 for x and 0-8 for y)
plt.show()
```

```python
x = np.arange(3) #same as before
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.xlim([0, 2])
plt.ylim([0, 8])
plt.show()
```

```python
x = np.arange(3) #Labels 
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.xlabel('x = arange(3)')
plt.xlim([0, 2])
plt.ylabel('y = f(x)')
plt.ylim([0, 8])
plt.show()
```

```python
x = np.arange(3) #Title 
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.grid(True)
plt.xlabel('x = arange(3)')
plt.xlim([0, 2])
plt.ylabel('y = f(x)')
plt.ylim([0, 8])
plt.title('Simple Plot Demo')
plt.show()
```

```python
x = np.arange(3) #Legends
plt.plot(x, x**2, label = 'x^2')
plt.plot(x, x**3, label = 'x^3')
plt.plot(x, 2*x, label = '2*x')
plt.plot(x, 2**x, label = '2^x')
plt.legend()
plt.grid(True)
plt.xlabel('x = arange(3)')
plt.xlim([0, 2])
plt.ylabel('y = f(x)')
plt.ylim([0, 8])
plt.title('Simple Plot Demo')
plt.show()
```

```python
x = np.arange(3) #Same as before
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.legend(['x^2', 'x^3', '2*x', '2^x'])
plt.grid(True)
plt.xlabel('x = arange(3)')
plt.xlim([0, 2])
plt.ylabel('y = f(x)')
plt.ylim([0, 8])
plt.title('Simple Plot Demo')
plt.show()
```

```python
x = np.arange(3) #Change position of the legend
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.legend(['x^2', 'x^3', '2*x', '2^x'], loc = 'upper center')
plt.grid(True)
plt.xlabel('x = arange(3)')
plt.xlim([0, 2])
plt.ylabel('y = f(x)')
plt.ylim([0, 8])
plt.title('Simple Plot Demo')
plt.show()
```

```python
x = np.arange(3) #Save the plot on a file
plt.plot(x, x**2, x, x**3, x, 2*x, x, 2**x)
plt.legend(['x^2', 'x^3', '2*x', '2^x'], loc = 'upper center')
plt.grid(True)
plt.xlabel('x = arange(3)')
plt.xlim([0, 2])
plt.ylabel('y = f(x)')
plt.ylim([0, 8])
plt.title('Simple Plot Demo')
plt.savefig('test.png')
plt.show()
```
