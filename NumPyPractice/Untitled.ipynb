# Bitwise Operations

```python
import numpy as np

x = np.array([0, 1, 0, 1], np.uint8)
y = np.array([0, 0, 1, 1], np.uint8)

print(np.bitwise_and(x, y)) #AND operations between two vectors (same size)
```

```python
print(np.bitwise_or(x, y)) #OR operations between two vectors
print(np.bitwise_xor(x, y)) #AND operations between two vectors
print(np.bitwise_not(y)) #NOT operation for "y" vector the results shows the complement of and 8 bit 0 and 00000001
```
