# Statistical Functions

```python
import numpy as np

a = np.arange(10)
print(a)
```

```python
print(np.median(a)) #mediana
```

```python
print(np.average(a)) #promedio
```

```python
print(np.mean(a)) #promedio
```

```python
b = np.random.randint(low = 0, high = 10, size = 10)
print(b)
```

```python
print(np.median(b))
print(np.average(b))
print(np.mean(b)) 
```

```python
print(np.std(b)) #Desviacion estandar
```

```python
print(np.var(b)) #Varianza
```

```python
print(np.histogram(b))
```
