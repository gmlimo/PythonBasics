# Arrays and Matplotlib
One and Zeros

```python
import numpy as np

x = np.empty([3, 3], np.uint8) #And array with random values
print(x)
```

```python
y = np.eye(3, dtype = np.uint8) #Matriz identidad
print(y)
```

```python
y = np.eye(5, dtype = np.uint8, k = 1) #Matriz identidad desplazada a la derecha
print(y)
```

```python
y = np.eye(5, dtype = np.uint8, k = -1) #Matriz identidad desplazada a la izquierda
print(y)
```

```python
x = np.identity(5, dtype = np.uint8)
print(x)
```

```python
x = np.ones((2, 5, 5), dtype = np.int16) #Two matrix with 5x5 dimension filled with ones
print(x)
```

```python
x = np.zeros((2, 5, 5), dtype = np.int16) #Two matrix with 5x5 dimension filled with zeros
print(x)
```

```python
x = np.full((3, 3, 3), dtype = np.int16, fill_value = 7) #Three matrix with 3x3 dimension filled with sevens
print(x)
```

# Matrix creation routines

```python
x = np.tri(3, 3, k = 0, dtype = np.uint16)
print(x)
```

```python
x = np.tri(5, 5, k = 1, dtype = np.uint16)
print(x)
```

```python
x = np.tri(5, 5, k = -1, dtype = np.uint16)
print(x)
```

```python
x = np.ones((5,5), dtype = np.uint8)
y = np.tril(x, k = -1) #Triangular matrix with the lower values filled
print(y)
```

```python
x = np.ones((5,5), dtype = np.uint8)
y = np.triu(x, k = 0) #Triangular matrix with the upper values filled
print(y)
```
