# Array Manipulation

```python
import numpy as np

x = np.arange(6)
print(x)
```

```python
y = x.reshape(3, 2) #Reshape the vector x as a 3x2 matrix
print(y)
```

```python
x = np.array([[0, 1, 2], [3, 4, 5]], dtype = np.uint8)
print(x)
```

```python
y = np.reshape(x, 6) #same as y = x.reshape(1, 6)
print(y)
```

```python
y = np.ravel(x)
print(y)
```

```python
y = x.flatten()
print(y)
```

```python
y = x.flatten('C') #Flatten ordered by rows
print(y)
```

```python
y = x.flatten('F') #Flatten ordered by columns
print(y)
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(x.dtype) #You can display the default data type of the arrays

print(x)
print(y)
```

```python
x = np.array([1, 2, 3], dtype = np.uint8)
y = np.array([4, 5, 6], dtype = np.uint8)

print(x.dtype) 

print(x)
print(y)
```

```python
z = np.stack((x, y))
print(z)
```

```python
z = np.stack((x, y), axis = 1)
print(z)
```

```python
z = np.dstack((x, y)) #same as before
print(z)
```

```python
print(z.shape)
```

```python
z = np.hstack((x, y)) #stack everything horizontally
print(z)
```

```python
z = np.vstack((x, y)) #same as the first stack
print(z)
```

```python
print(z.shape)
```

```python
x = np.arange(9)
print(x)
```

```python
a, b, c = np.split(x, 3) #splits x in 3 separated arrays
print(a, b, c)
```

```python
x = np.random.rand(4, 4, 4) #generates a random array of 4 dimensions with the size of 4x4
print(x)
```

```python
y, z = np.split(x, 2)
print(y, z)
```

```python
x = np.arange(16).reshape(4, 4)
print(x)
```

```python
y = np.flip(x) #Inverts all the contents of the array
print(y)
```

```python
y = np.flip(x, axis = -1) #Inverts all the contents of the array like a mirror
print(y)
```

```python
y = np.fliplr(x) #same as before
print(x)
print(y)
```

```python
y = np.flipud(x) #Up and down
print(x)
print(y)
```

```python
y = np.roll(x, 8) #A rotation of 8 elements
print(y)
```

```python
y = np.rot90(x)
print(y)
```
